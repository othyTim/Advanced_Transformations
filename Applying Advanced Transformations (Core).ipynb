## Imports

```python
## Standard Imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
## Importing the OS and JSON Modules
import os,json
```

```python
super_info = pd.read_csv('Data/superhero_info - superhero_info.csv')
super_info.info()
super_info.head()
```

```python
super_power = pd.read_csv('Data/superhero_powers - superhero_powers.csv')
super_power.info()
super_power.head()
```

## Data Cleanup

### Hero|Publisher Column

```python
super_info['Hero|Publisher'].head(2)
```

```python
# Splitting Hero and Publisher
super_info['Hero|Publisher'].str.split('|',expand=True)
```

```python
# Saving the 2 columns into our DataFrame
super_info[['Hero','Publishers']] = super_info['Hero|Publisher'].str.split('|',expand=True)
super_info.head(2)
```

```python
# Drop original Column 
super_info = super_info.drop(columns=['Hero|Publisher'])
super_info.head(2)
```

### Measurements Column

```python
## examining a single value from the coordinates col
meas = super_info.loc[0,"Measurements"]
print(type(meas))
meas
```

```python
# replace single quotes with double quotes before we use json.loads
meas = meas.replace("'",'"')
meas
```

```python
## now we can use json.loads
fixed_meas = json.loads(meas)
print(type(fixed_meas))
fixed_meas
```

```python
## use .str.replace to replace all single quotes
super_info['Measurements'] = super_info['Measurements'].str.replace("'",'"')
## Apply the json.loads to the full column
super_info['Measurements'] = super_info['Measurements'].apply(json.loads)
super_info['Measurements'].head()
```

```python
## check a single value after transformation
test_meas = super_info.loc[0, 'Measurements']
print(type(test_meas))
test_meas
```

```python
# Unpack a column of dictionaries into separate columns
hei_wei = super_info['Measurements'].apply(pd.Series)
hei_wei
```

```python
# concat hei_wei with original dataframe
super_info = pd.concat((super_info, hei_wei), axis = 1)
super_info.head(2)
```

```python
# Drop original measurement column
super_info = super_info.drop(columns=['Measurements'])
```

```python
super_info.head(2)
```

```python
super_info.info()
```

```python
# rename Columns to include measuremnt in order to remove it from the fields
super_info.rename(
    columns={"Height": "Height(cm)", "Weight": "Weight(kg)"},
    inplace=True,
)
super_info.head(2)
```

```python
super_info['Height(cm)'] = super_info['Height(cm)'].str.replace(' cm', '')
super_info['Weight(kg)'] = super_info['Weight(kg)'].str.replace(' kg', '')
```

```python
super_info.head(2)
```

```python
# change column type to numeric
super_info = super_info.astype({'Height(cm)':'float'})
super_info = super_info.astype({'Weight(kg)':'float'})
```

```python
super_info.info()
```

## One Hot Encoding Powers

```python
super_power.head(2)
```

```python
# Split the "Powers" column to extract individual powers and create a list of unique powers
all_powers = set(','.join(super_power['Powers']).split(','))

# Create one-hot-encoded columns for each unique power
for power in all_powers:
    super_powers = super_power['Powers'].str.get_dummies(sep=',')
super_power = pd.concat([super_power, super_powers], axis=1)

# Drop the original "Powers" column
super_power.drop(columns=['Powers'], inplace=True)

# Print the resulting DataFrame
super_power
```

```python
super_power.head()
```

## Merge DataFrames

```python
#merge with our 2 dataframes
final_heros = pd.merge(super_info, super_power, left_on='Hero', right_on='hero_names', how='inner')
final_heros.head(3)
```

## Info about the Super Heros

```python
# Create a subset of the DataFrame for superpowers with Super Speed
super_speed_df = final_heros[final_heros['Super Speed'] == True]

# Create a subset of the DataFrame for superpowers without Super Speed
no_super_speed_df = final_heros[final_heros['Super Speed'] == False]

# Calculate the average weight for each subset
average_weight_super_speed = super_speed_df['Weight(kg)'].mean()
average_weight_no_super_speed = no_super_speed_df['Weight(kg)'].mean()

# Print the results
print("Average weight of superpowers with Super Speed:", average_weight_super_speed)
print("Average weight of superpowers without Super Speed:", average_weight_no_super_speed)
```

```python
# Group the data by 'Publishers' and calculate the average height for each publisher
average_height_by_publisher = final_heros.groupby('Publishers')['Height(cm)'].mean()

average_height_by_publisher
```
